```python
#G: Genuino
#I: No es genuino
numero_de_grabaciones=int(input("Ingrese el n√∫mero de grabaciones disponibles: "))

claves_genuino=[]
claves_no_genuino=[]

for grabaciones in range(numero_de_grabaciones):
  claves_genuino.append("G"+str(grabaciones+1))
  claves_no_genuino.append("I"+str(grabaciones+1))
print("")
print(claves_no_genuino)
print(claves_genuino)

import random

probabilidad_similaridad_genuino=[]
probabilidad_tipicalidad_genuino=[]
probabilidad_similaridad_nogenuino=[]
probabilidad_tipicalidad_nogenuino=[]
grabaciones_genuino={}
grabaciones_no_genuino={}

for probabilidad in range(numero_de_grabaciones):
  probabilidad_similaridad_genuino.append(random.random())
  probabilidad_tipicalidad_genuino.append(random.random())
  probabilidad_similaridad_nogenuino.append(random.random())
  probabilidad_tipicalidad_nogenuino.append(random.random())

i=0
for grabaciones in range(numero_de_grabaciones):
  grabaciones_genuino[claves_genuino[i]]=probabilidad_similaridad_genuino[i]/probabilidad_tipicalidad_genuino[i]
  grabaciones_no_genuino[claves_no_genuino[i]]=probabilidad_similaridad_nogenuino[i]/probabilidad_tipicalidad_nogenuino[i]
  i=i+1

print("")
print(probabilidad_similaridad_genuino)
print(probabilidad_tipicalidad_genuino)
print(probabilidad_similaridad_nogenuino)
print(probabilidad_tipicalidad_nogenuino)
print("")
print("Likeihood ratio: ")
print(grabaciones_genuino)
print(grabaciones_no_genuino)

import matplotlib.pyplot as plt

grabaciones=list(grabaciones_genuino.keys())+list(grabaciones_no_genuino.keys())
probabilidades=list(grabaciones_genuino.values())+list(grabaciones_no_genuino.values())

plt.bar(grabaciones,probabilidades)

plt.xlabel('Grabaciones')
plt.ylabel('Score')
plt.title('Histograma grabaciones genuinas y no genuinas')

# Muestra el histograma
plt.show()
```
