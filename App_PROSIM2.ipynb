```python
import tkinter as tk
from tkinter import ttk
#########################################################################################################################
ventana = tk.Tk()
ventana.geometry('1000x1000')
ventana.title('Menú principal')
ventana.iconbitmap('C:/Users/josea/Documents/Servicio social/ProSim2.ico')
########################################################################################################################
style = ttk.Style()# Creamos un estilo personalizado
style.configure('TButton', foreground='black', background='lightblue', font=('Arial', 12))
style.configure('TLabel', foreground='darkblue', background='white', font=('Arial', 14), justify='right')
########################################################################################################################
etiqueta1 = ttk.Label(ventana, text='     ')
etiqueta1.grid(row=10, column=10, columnspan=2)
#########################################################################################################################
def ECG():
    etiqueta1.config(text="Usted eligió: ECG onda")
    ventana_ECG = tk.Toplevel(ventana)
    ventana_ECG.geometry('800x800')
    ventana_ECG.title('ECG onda')
    
    aceptar=ttk.Button(ventana_ECG, text='Aceptar')
    aceptar.grid(row=20, column=25)
    
    etiqueta_ECG=ttk.Label(ventana_ECG, text='Tasa ECG (BPM): ')
    etiqueta_amplitud=ttk.Label(ventana_ECG, text='Amplitud (mV): ')
    etiqueta_tipo=ttk.Label(ventana_ECG, text='Tipo (Adulto/Niño): ')
    etiqueta_st=ttk.Label(ventana_ECG, text='ST: ')
    etiqueta_artefacto=ttk.Label(ventana_ECG, text='Artefacto (On/Off): ')
    #///////////////////////////////////////////////////////////////////
    etiqueta_ECG.grid(row=10, column=22, columnspan=2)
    etiqueta_ECG.configure(style='TLabel')
    
    etiqueta_amplitud.grid(row=11, column=22, columnspan=2)
    etiqueta_amplitud.configure(style='TLabel')
    
    etiqueta_tipo.grid(row=12, column=22, columnspan=2)
    etiqueta_tipo.configure(style='TLabel')
    
    etiqueta_st.grid(row=13, column=22, columnspan=2)
    etiqueta_st.configure(style='TLabel')
    
    etiqueta_artefacto.grid(row=14, column=22, columnspan=2)
    etiqueta_artefacto.configure(style='TLabel')
    #///////////////////////////////////////////////////////////////////
    entrada_ECG=ttk.Entry(ventana_ECG, width=40, textvariable='Ingresa los latidos por minuto: ')
    entrada_ECG.grid(row=10, column=25)
    
    entrada_amplitud=ttk.Entry(ventana_ECG, width=40, textvariable='Ingresa la amplitud: ')
    entrada_amplitud.grid(row=11, column=25)
    
    entrada_tipo=ttk.Entry(ventana_ECG, width=40, textvariable='Ingresa el tipo: ')
    entrada_tipo.grid(row=12, column=25)
    
    entrada_st=ttk.Entry(ventana_ECG, width=40, textvariable='Ingresa el ST: ')
    entrada_st.grid(row=13, column=25)
    
    entrada_artefacto=ttk.Entry(ventana_ECG, width=40, textvariable='Ingresa el artefacto: ')
    entrada_artefacto.grid(row=14, column=25)
      
########################################################################################################################
def arry():
    etiqueta1.config(text="Usted eligió: Arritmia")
    ventana_arry=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_arry.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_arry.title('Arritmia')
########################################################################################################################    
def temp():
    etiqueta1.config(text="Usted eligió: Temperatura")
    ventana_temp=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_temp.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_temp.title('Temperatura')
########################################################################################################################    
def resp():
    etiqueta1.config(text="Usted eligió: Respiración")
    ventana_resp=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_resp.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_resp.title('Respiración')
    
def perf():
    etiqueta1.config(text="Usted eligió: Onda de rendimiento")
    ventana_perf=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_perf.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_perf.title('Onda de rendimiento')
########################################################################################################################    
def co():
    etiqueta1.config(text="Usted eligió: Gasto cardiaco")
    ventana_co=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_co.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_co.title('Señal de salida de gasto cardiaco')
########################################################################################################################    
def bp():
    etiqueta1.config(text="Usted eligió: Presión sanguínea")
    ventana_bp=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_bp.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_bp.title('Presión sanguínea')
######################################################################################################################## 
def pace():
    etiqueta1.config(text="Usted eligió: Marcapasos")
    ventana_pace=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_pace.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_pace.title('Marcapasos')
########################################################################################################################    
def fema():
    etiqueta1.config(text="Usted eligió: Fetal/Materno")
    ventana_fema=tk.Tk()
    #Modificamos el tamaño de la ventana (píxeles)
    ventana_fema.geometry('700x700')
    #Cambiamos el nombre de la ventana
    ventana_fema.title('Fetal / Materno')
######################################################################################################################## 
def aceptar():
    print("Aceptar")
########################################################################################################################

ECG_onda= ttk.Button(ventana, text='ECG Onda', command=ECG)
ECG_onda.grid(row=1, column=1)
########################################################################################################################
arritmia=ttk.Button(ventana, text='Arritmia', command=arry)
arritmia.grid(row=2, column=1)
########################################################################################################################
temperatura=ttk.Button(ventana, text='Temperatura', command=temp)
temperatura.grid(row=3, column=1)
########################################################################################################################
respiracion=ttk.Button(ventana, text='Respiración', command=resp)
respiracion.grid(row=1, column=2)
########################################################################################################################
onda_de_rendimiento=ttk.Button(ventana, text='Onda de rendimiento', command=perf)
onda_de_rendimiento.grid(row=2, column=2)
########################################################################################################################
gasto_cardiaco=ttk.Button(ventana, text='Gasto cardiaco', command=co)
gasto_cardiaco.grid(row=3, column=2)
########################################################################################################################
presion_sanguinea=ttk.Button(ventana, text='Presión sanguinea', command=bp)
presion_sanguinea.grid(row=1, column=3)
########################################################################################################################
marca_pasos=ttk.Button(ventana, text='Marcapasos', command=pace)
marca_pasos.grid(row=2, column=3)
########################################################################################################################
fetal_materno=ttk.Button(ventana, text='Fetal / Materno', command=fema)
fetal_materno.grid(row=3, column=3)
########################################################################################################################

# Aplicamos el estilo a los botones
ECG_onda.configure(style='TButton')
arritmia.configure(style='TButton')
temperatura.configure(style='TButton')
respiracion.configure(style='TButton')
onda_de_rendimiento.configure(style='TButton')
gasto_cardiaco.configure(style='TButton')
presion_sanguinea.configure(style='TButton')
marca_pasos.configure(style='TButton')
fetal_materno.configure(style='TButton')
########################################################################################################################
# Aplicamos el estilo a las etiquetas
etiqueta1.configure(style='TLabel')
########################################################################################################################
ventana.mainloop()
```
